**Image to Pencil Sketch with Python**

---



---



```python
#import necessary libraries
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
def convert_to_pencil_sketch(image_path, output_path):
    try:
        # Read the image in RGB format
        image = cv2.imread('/content/i 1.jpg')

        # Convert the image to grayscale
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Invert the grayscale image
        inverted_image = cv2.bitwise_not(gray_image)

        # Apply Gaussian blur to the inverted image
        blurred_image = cv2.GaussianBlur(inverted_image, (111, 111), 0)

        # Divide the grayscale image by the inverted blurred image
        sketch = cv2.divide(gray_image, blurred_image, scale=256.0)

        # Display the sketch
        cv2_imshow(sketch)

        # Save the sketch to a file
        cv2.imwrite(output_path, sketch)

        print("Pencil sketch saved successfully as", output_path)
    except Exception as e:
        print("An error occurred:", str(e))
```

```python
# Example usage
image_path = "your_image.jpg"
output_path = "pencil_sketch.jpg"
convert_to_pencil_sketch(image_path, output_path)
```
